## טיפול בנתונים

```python
#בדיקת כפילויות
#קיימות 3 כפילויות
df.duplicated().sum()
```

```python
#הורדת עמודת הכפילויות
df = df.drop_duplicates()
```

```python
# - Medication_Use מציאת הנתון השכיח בעמודה
commonValueMed = df["Medication Use"].mode()[0]
```

```python
# - Substance_Use מציאת הנתון השכיח בעמודה
commonValueSub = df["Substance Use"].mode()[0]
```

```python
#מילוי הנתונים החסרים בנתון השכיח ביותר 
df["Medication Use"]= df["Medication Use"].fillna(commonValueMed)
```

```python
#מילוי הנתונים החסרים בנתון השכיח ביותר 
df["Substance Use"]= df["Substance Use"].fillna(commonValueSub)
```

```python
#מחיקת כל השורות עם הנתונים החסרים
df=df.dropna()
```

```python
#העמודה של ההיסטוריה המשפחתית כרגע מוגדת לfloat.
#שינוי העמודה לint
df['Family History Mental Illness'] = df['Family History Mental Illness'].astype(int)
```

```python
#שינוי הספרות 0ו-1 לצירוף מילים ברור יותר.
df['Family History Mental Illness'] = df['Family History Mental Illness'].replace([0,1],["No Family History","Yes Family History"])
```

```python
df['Family History Mental Illness']
```

```python
#שינוי הספרות 0ו-1 לצירוף מילים ברור יותר.
df['Chronic Illnesses'] = df['Chronic Illnesses'].replace([0,1],["No","Yes"])
```

```python
df['Chronic Illnesses']
```

```python
#מחיקה של עמודה
#עמודה פחות רלוונטית למחקר שלנו לכן נמחק אותה
df.drop(['Social Support Score'],axis = 1, inplace = True)
```

```python
#הוספת עמודה Sleep_Category
#חלוקה ל3 קבוצות עיקריות
#לצורך הצגה וניתוח ברור של הנתונים
# פונקציה פשוטה לעמודת Sleep_Hours
def sleep_category(hours):
    if hours < 4:
        return 'Low'
    elif hours < 7:
        return 'Medium'
    else:
        return 'High'

df['Sleep Hours Level'] = df['Sleep Hours'].apply(sleep_category)

```

```python
#הוספת עמודה Physical_activity_category
#חלוקה ל3 קבוצות עיקריות
#לצורך הצגה וניתוח ברור של הנתונים
def physical_activity_category(hours):
    if hours < 0.7:
        return 'Low Activity'
    elif hours < 1.4:
        return 'Medium Activity'
    else:
        return 'High Activity'

df['Physical Activity Level'] = df['Physical Activity Hrs'].apply(physical_activity_category)
    
```

* קיים קשר חזק מרוד בין חרדה לדיכאון המצביע על כך שמדובר במימדים חופפים בחווית המצוקה של הנבדקים במדגם.
* נמצא קשר שלילי חזק בין מספר שעות השינה לרמות חרדה ודיכאון. ככל שישנים יותר כך רמת החרדה ודכאון פוחתת.
* פעילות גופנית נמצאה קשורה לרמות נמוכות של חרדה ודיכאון - ככל שהתדירות עולה, המצוקה הנפשית פוחתת.
* משתתפים שמשלבים טיפול פסיכולוגי עם מדיטציה מדווחים על רמות דיכאון נמוכות יותר ממי שלא משתמש באמצעים או.
* קבוצת הגיל 20-30 דיווחו על רמות דיכאון גבוהות מהממוצע. קבוצת הגיל 30-40 נמצאה במקום השני עם ציון נמוך במעט.
* הנתונים מצביעים על קשר מובהק בין מצב תעסוקתי לרמות חרדה ודיכאון, כאשר משתתפים בלתי מועסקים מדווחים על רמות גבוהות משמעותית של מצוקה נפשית, בהשוואה למועסקים ולסטודנטים.
הפערים בולטים במיוחד בקרב Non-Binary ונשים
* נתוני המדגם מראים כי שימוש מתון בחומרים (כגון אלכוהול או חומרים דמויי סם) אינו קשור לעלייה ברמות החרדה והדיכאון, ולעיתים אף מופיעים בקרב משתמשים מזדמנים ציונים מעט נמוכים יותר לעומת אלו שנמנעו לחלוטין. עם זאת, שימוש תדיר נמצא קשור לעלייה מובהקת ברמות המצוקה הנפשית.
* מחלות כרוניות נמצאו קשורות לעלייה קלה ברמות חרדה ודיכאון, בעוד שהיסטוריה משפחתית של מחלה נפשית לא הציגה הבדל מובהק במדדים אלו.
